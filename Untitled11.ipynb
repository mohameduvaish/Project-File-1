```python
import pandas as pd

# Update with your file path in Drive
df = pd.read_csv('/content/drive/MyDrive/credit_risk.csv')

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Replace 'target' with your dataset label column
X = df.drop(['target', 'loan_purpose'], axis=1)
y = df['target']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Scaling numeric features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

# XGBoost Model
xgb = XGBClassifier(n_estimators=300, max_depth=5, learning_rate=0.05)
xgb.fit(X_train_scaled, y_train)
xgb_pred = xgb.predict_proba(X_test_scaled)[:,1]
xgb_auc = roc_auc_score(y_test, xgb_pred)

# Random Forest Model
rf = RandomForestClassifier(n_estimators=300, max_depth=10)
rf.fit(X_train, y_train)
rf_pred = rf.predict_proba(X_test)[:,1]
rf_auc = roc_auc_score(y_test, rf_pred)

print("XGBoost AUC:", xgb_auc)
print("Random Forest AUC:", rf_auc)
```

```python
best_model = xgb if xgb_auc >= rf_auc else rf
print("Best model selected:", type(best_model).__name__)
```

```python
import shap
import matplotlib.pyplot as plt

explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X_test)

# Summary plot
shap.summary_plot(shap_values, X_test, plot_type="dot")
plt.savefig('/content/drive/MyDrive/shap_summary_plot.png')
```

```python
import numpy as np

# Get top 3 important features
feature_importance = np.mean(np.abs(shap_values), axis=0)
top_features = X.columns[np.argsort(feature_importance)[-3:]]

for feat in top_features:
    shap.dependence_plot(feat, shap_values, X_test)
```

```python
# Identify example cases
high_risk_index = X_test.index[y_test == 1][0]
low_risk_index = X_test.index[y_test == 0][0]

# Force plot high risk
shap.force_plot(explainer.expected_value,
                shap_values[X_test.index.get_loc(high_risk_index)],
                X_test.loc[high_risk_index],
                matplotlib=True)

# Force plot low risk
shap.force_plot(explainer.expected_value,
                shap_values[X_test.index.get_loc(low_risk_index)],
                X_test.loc[low_risk_index],
                matplotlib=True)
```

```python
# Save SHAP values
pd.DataFrame(shap_values).to_csv(
    '/content/drive/MyDrive/shap_values_output.csv', index=False
)
print("All results saved to Google Drive.")
```
